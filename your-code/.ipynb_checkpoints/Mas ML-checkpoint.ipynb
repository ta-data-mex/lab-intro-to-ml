```python
#%% 
import pandas, numpy
import matplotlib.pyplot as plt
import copy
%matplotlib inline
```

```python
#%%
df = pandas.DataFrame({
    'x': [12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 55, 53, 55, 61, 64, 69, 72],
    'y': [39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58, 23, 14, 8, 19, 7, 24]
})
```

```python
#%%
numpy.random.seed(200)
k = 3
centroides = {
    i+1: [numpy.random.randint(0,80), numpy.random.randint(0,80)]
    for i in range(k)}
```

```python
#%%
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'], df['y'], color='k')
colmap= {1: 'r', 2: 'g', 3:'b'}

for i in centroides.keys():
    plt.scatter(*centroides[i], color= colmap[i])
plt.xlim(0,80)
plt.ylim(0,80)
plt.show()
```

```python
#%%
def asignacion(df, centroides):
    for i in centroides.keys():
        df['distance_from_{}'.format(i)] =(
            numpy.sqrt(
                (df['x'] - centroides[i][0]) **2 +(df['y']-centroides[i][1])**2
            )
        )
    centroides_distance_cols=['distance_from_{}'.format(i) for i in centroides.keys()]
    df['closest'] = df.loc[:,centroides_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df
```

```python
df = asignacion(df, centroides)
```

```python
#%%
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor='k')
for i in centroides.keys():
    plt.scatter(*centroides[i], color=colmap[i])
plt.ylim(0,80)
plt.xlim(0,80)
plt.show()
```

```python
#%%
old_centroides = copy.deepcopy(centroides)
```

```python
#%%
def actualizar(k):
    for i in centroides.keys():
        centroides[i][0] = numpy.mean(df[df['closest'] ==i]['x'])
        centroides[i][1] = numpy.mean(df[df['closest'] ==i]['y'])
    return k
```

```python
#%%
centroides = actualizar(centroides)
```

```python
#%%
fig = plt.figure(figsize=(5,5))
ax = plt.axes()
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.5, edgecolor = 'k')
for i in centroides.keys():
    plt.scatter(*centroides[i], color=colmap[i])
plt.xlim(0,80)
plt.ylim(0,80)
for i in old_centroides.keys():
    old_x = old_centroides[i][0]
    old_y = old_centroides[i][1]
    dx = (centroides[i][0] - old_centroides[i][0]) * 0.75
    dy = (centroides[i][1] - old_centroides[i][1]) * 0.75
    ax.arrow(old_x, old_y, dx, dy, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
plt.show()
```

```python
#%%
df = asignacion(df, centroides)
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'], df['y'], color= df['color'], alpha=0.5, edgecolor='k')

for i in centroides.keys():
    plt.scatter(*centroides[i], color = colmap[i])
plt.xlim(0,80)
plt.ylim(0,80)
plt.show()
```

```python
while True:
    closest_centroides=df['closest'].copy(deep=True)
    centroides = actualizar(centroides)
    df = asignacion(df,centroides)
    if closest_centroides.equals(df['closest']):
        break
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],color = df['color'],alpha=0.5,edgecolor = 'k')

for i in centroides.keys():
    plt.scatter(*centroides[i], color=colmap[i])
plt.xlim(0,80)
plt.ylim(0,80)
plt.show()
```
